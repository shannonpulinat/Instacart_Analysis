# 4 Create Loyalty Flag 

```python
df_ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge.head(20)
```

```python
# Check frequencies of loyalty flags 
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 5. mean prices of products for each type of customer

```python
# Group by "prices" 
df_ords_prods_merge.groupby('prices')
```

```python
# Split  data into groups by "loyalty_flag" and  calculate the mean for each "prices" column
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

# 6. Create Spending Flag based on avg price across all orders

```python
# Group data by "user_id" column
#generate mean product price for each user
#create "mean_price" for aggregation results
df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head(20)
```

```python
# Create flags to assign "spending" labels to user ID based 
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Check frequency
df_ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

# 7 Create order frequency flag based on days_since_prior_order


```python
# Group data by "user_id" column
#generate median days for each user
#create "median_days" for aggregation results
df_ords_prods_merge['median_days'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Check Group
df_ords_prods_merge.head(20)
```

```python
#Create flags 
df_ords_prods_merge.loc[df_ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days'] > 10) & (df_ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
